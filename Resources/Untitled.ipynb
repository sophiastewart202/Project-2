```python
import pandas as pd
from sqlalchemy import create_engine, Table, Column, Integer, MetaData
import numpy as np
from sqlalchemy.ext.automap import automap_base
```

```python
filepath = "../clean_reviews_data.db"
engine = create_engine(f"sqlite:///{filepath}")
data_df = pd.read_sql('SELECT AUTHOR, TITLE, REVIEW, RATING FROM "IMDB_Reviews"', engine)
data_df
```

```python
data_df["AUTHOR_ID"] = data_df.index +1
data_df
```

```python
data_df.to_csv("IMDB_Reviews.csv", index = False)
```

```python
author_data = data_df[["AUTHOR_ID","AUTHOR"]]
author_data.set_index('AUTHOR_ID')
author_data =author_data.sort_index(ascending = True)
author_data
```

```python
author_data.to_csv("authors.csv", index = False)
```
