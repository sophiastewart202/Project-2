```python
import pandas as pd
import numpy as np

from sqlalchemy import create_engine
from sqlalchemy import Table, Column, Integer, String, MetaData
metadata = MetaData()
```

```python
filepath = "Resources/clean_reviews_data.db"
engine = create_engine(f"sqlite:///{filepath}")
df = pd.read_sql('SELECT AUTHOR, TITLE, REVIEW, RATING FROM "IMDB Reviews"', engine)
df
```

```python
imdb = df.copy()
imdb["AUTHOR_ID"] = imdb.index + 1
imdb
```

```python
# imdb.filter["AUTHOR_ID, AUTHOR"]
author = imdb[["AUTHOR_ID", "AUTHOR"]]
author
```

```python
imdb["AUTHOR"].nunique()
```

```python
num_reviews = imdb.sort_values('AUTHOR', ascending=True).groupby("AUTHOR")["TITLE"].count()
num_reviews_df = pd.DataFrame(num_reviews)
top_reviewers = num_reviews_df.sort_values("TITLE", ascending=False).head(20)
top_reviewers
```
