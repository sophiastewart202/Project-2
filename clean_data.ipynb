```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
filepath = "Resources/IMDB_Movies_2021.db"
engine = create_engine(f"sqlite:///{filepath}")
df = pd.read_sql("SELECT ID, AUTHOR, TITLE, REVIEW, RATING FROM REVIEWS", engine)
df
```

```python
null_counts = df['RATING'].isnull()
null_counts

clean_df = df.dropna().copy()
clean_df
```

```python
clean_df['TITLE'].replace('\n', '', regex=True, inplace=True)
clean_df['REVIEW'].replace('\n', '', regex=True, inplace=True)
clean_df
```

```python
engine = create_engine('sqlite:///clean_reviews_data.db', echo=True)
sqlite_connection = engine.connect()
```

```python
sqlite_table = "IMDB_Reviews"
clean_df.to_sql(sqlite_table, sqlite_connection, if_exists='fail')
```

```python
sqlite_connection.close()
```
