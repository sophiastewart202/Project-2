```python
# Dependencies and Setup
import pandas as pd

# File location
src_path = 'Resources/aggr/'
filelist = {'VAERSDATA.csv': [1,7,10,18,19,30], 'VAERSSYMPTOMS.csv': [], 'VAERSVAX.csv': []}

# Indentation for reporting
sp4 = '    '
sp8 = '        '

# Process each file and list properties of each column
for file in filelist.keys():
    print(f"*******************************\nFilename: {file}")
    # Set 'VAERS_ID' as index column. This also confirms that 'VAERS_ID' is unique (should get error if not)
    df = pd.read_csv(src_path + file, index_col='VAERS_ID', parse_dates=filelist[file], low_memory=False)
    print(f"{sp4}Record count: {len(df):,}\n*******************************")
    for col in df.columns:
        print()
        print(f"Column name: {col}")
        print(f"{sp4}dtype: {df[col].dtype}")
        if df[col].dtype == 'datetime64[ns]':
            print(f"{sp8}Min Date: {df[col].min()}")
            print(f"{sp8}Max Date: {df[col].max()}")
            num_null = df[col].isna().sum()
            print(f"{sp8}Null count: {num_null:,}    percent: {num_null/len(df):.2%}")
        elif df[col].dtype == 'float64':
            print(f"{sp8}Mean: {df[col].mean()}")
            print(f"{sp8}Minimum: {df[col].min()}")
            print(f"{sp8}1st quartile: {df[col].quantile(0.25)}")
            print(f"{sp8}Median: {df[col].median()}")
            print(f"{sp8}3rd quartile: {df[col].quantile(0.75)}")
            print(f"{sp8}Maximum: {df[col].max()}")
        else:
            s_report = df[col].value_counts(normalize=True, dropna=False).head(10)
            s_report['Remaining'] = 1 - s_report.sum()
            s_report = s_report.map("{:.2%}".format)
            rows = (s_report.to_string())
            for row in rows.splitlines():
                print(f"{sp8}{row}")
    print()
                
```
