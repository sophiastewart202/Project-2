```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
filepath = "Resources/IMDB_Movies_2021.db"
engine = create_engine(f"sqlite:///{filepath}")
df = pd.read_sql("SELECT AUTHOR, TITLE, REVIEW, RATING FROM REVIEWS", engine)
df
```

```python
null_counts = df['RATING'].isnull()
null_counts

clean_df = df.dropna().copy()
clean_df
```

```python
clean_df['TITLE'].replace('\n', '', regex=True, inplace=True)
clean_df['REVIEW'].replace('\n', '', regex=True, inplace=True)
clean_df['REVIEW'].replace(',', '',regex = True, inplace = True)
clean_df['REVIEW'].replace("'", '',regex = True, inplace = True)

clean_df
```

```python
clean_df = clean_df.drop_duplicates(subset='AUTHOR', keep='first')
clean_df
```

```python
clean_df.index = np.arange(1, len(clean_df) + 1)

clean_df["AUTHOR_ID"] = clean_df.index 

#clean_df = clean_df.pop('AUTHOR_ID')
clean_df
```

```python
engine = create_engine('sqlite:///clean_reviews_data.db', echo=True)
sqlite_connection = engine.connect()
```

```python
sqlite_table = "imdbreviews"
clean_df.to_sql(sqlite_table, sqlite_connection, if_exists='fail')
```

```python
clean_df.to_csv("imdbreviews.csv", index = False)
```

```python
author_data = clean_df[["AUTHOR_ID","AUTHOR"]]
author_data.set_index('AUTHOR_ID')
author_data =author_data.sort_index(ascending = True)
author_data
```

```python
author_data.to_csv("authors.csv", index = False)
```
